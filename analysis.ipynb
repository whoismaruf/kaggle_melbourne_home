# Import



```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

# Data wrangling

```python
df = pd.read_csv('melb_data.csv')
df.dropna(axis=0)
df.columns
```

# Features Engineering

```python
y = df.Price
features__columns = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = df[features__columns]
```

# Fitting Model

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
    model.fit(train_X, train_y)
    val_predictions = model.predict(val_X)
    return mean_absolute_error(val_y, val_predictions)
```

```python
max_leaf_nodes = [5, 25, 100, 500, 5000]
scores = {i:get_mae(i, train_X, val_X, train_y, val_y) for i in max_leaf_nodes}
best_tree_size = min(scores, key=scores.get)
best_tree_size
```

```python
final_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=1)
final_model.fit(train_X, train_y)
print("Training completed...!")
```
